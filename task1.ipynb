```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "weather_history_bangladesh.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
plt.figure(figsize=(25,15))

plt.hist(df.iloc[:, 4], bins=10, color='skyblue', alpha=0.7, label='Due_point')
plt.grid(True)
plt.hist(df.iloc[:, 10], bins=10, color='grey', alpha=0.7, label='Temprature')
plt.grid(True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('')
plt.legend()
plt.tight_layout()
plt.show()
```
